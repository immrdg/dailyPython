# **Frequency Distribution of each subject**

```python
def get_subject_hist(subject):
    fig,ax=plt.figure(),plt.axes()
    sub_hist = df[subject]
    sns.histplot(x=sub_hist,element='bars')
    plt.title(f"Frequency of {subject} subject.")
    plt.xlabel(f"{subject} marks.")
```

```python
get_subject_hist('Physics')
```

```python
df.columns[1:-3]
```

```python
def subjectsPlots():
    fig, axs = plt.subplots(ncols=2, nrows=3, figsize=(10, 10))
    index = 0
    axs = axs.flatten()
    for subject in df.columns[1:-3]:
        sns.boxplot(x=df[subject], data=df, ax=axs[index])
        index += 1
```

```python
subjectsPlots()
```

```python
def get_subject_hist1(subject):
    plt.subplots()
    sub_hist = df[subject]
    sns.histplot(x=sub_hist,element='bars')
    plt.title(f"Frequency of {subject} subject.")
    plt.xlabel(f"{subject} marks.")
```

```python
subjects = ['Physics', 'Chemistry', 'Mathematics', 'Biology','Computer Science', 'English',]
for subject in subjects:
    get_subject_hist1(subject)
    print("\n")
```

# **Giving Grades Based on Percentage**
### 1. 80 - 90 = A
### 2. 70 - 80 = B+
### 3. 60 - 70 = B
### 4. 50 - 60 = C+
### 5. 40 - 50 = D

```python
percentage = df['Percentage']
grades = [None for i in range(49)] #Because we have 49 students

def set_grade(minimum,maximum,grade):
    if minimum <= maximum:
        indexes = df[(percentage >minimum) & (percentage <=maximum)].index
        for index in indexes:
            grades[index] = grade
        print(grades,"\n")
    else:
        raise ValueError("maximum should be greater than minimum")
```

```python
set_grade(80,90,"A")
set_grade(70,80,"B+")
set_grade(60,70,"B")
set_grade(50,60,"C+")
set_grade(40,50,"C")
```

```python
df['Grade'] = grades
df
```

```python
def pieChart():
    plt.subplots()
    plt.style.use("fivethirtyeight")
    df['Grade'].value_counts().plot(kind='pie', autopct="%.2f",shadow=True)
    plt.title("Grades Pie Chart")
```

```python
pieChart()
```

```python
def StylepieChart():
    plt.subplots()
    plt.style.use("fivethirtyeight")
    df['Grade'].value_counts().plot(kind='pie', autopct="%.2f",shadow=True)
    plt.title("Grades Pie Chart")
StylepieChart()
```
